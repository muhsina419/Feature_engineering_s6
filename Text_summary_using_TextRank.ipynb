```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.text_rank import TextRankSummarizer
```

```python
text = """
Artificial Intelligence is transforming the way humans interact with technology.
Machine learning algorithms allow systems to learn from data and improve their performance over time.
Natural Language Processing enables machines to understand and generate human language.
These technologies are used in applications such as chatbots, recommendation systems, and autonomous vehicles.
AI is expected to significantly impact industries such as healthcare, finance, and education.
"""
```

```python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
parser = PlaintextParser.from_string(text, Tokenizer("english"))
```

```python
summarizer = TextRankSummarizer()
```

```python
summary = summarizer(parser.document, 2)
```

```python
for sentence in summary:
    print(sentence)
```
